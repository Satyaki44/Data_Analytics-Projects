```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("bestsellers with categories.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#what are count of differnt ratings?
plt.figure(figsize=(8,5))
sns.countplot(df["User Rating"].sort_values())
```

So we have found that 4.9 is the highest rating given by user.

```python
#what are the top rated books of 2019?
df2=df[df["Year"]==2019]
```

```python
df2.head()
```

```python
df3=df2[df2["User Rating"]==4.9]
```

```python
df3.head()
```

```python
df3["Name"]
```

```python
sns.countplot(df3["Name"],palette="winter")
plt.xticks(rotation=90)
```

#### Above are the names of the books that have highest ratings in 2019

```python
#distribution of fiction vs non fiction books
sns.countplot(df["Genre"])
```

```python
#Which Author has the maximum books of 4.9 ratings?
df4=df[df["User Rating"]==4.9]
plt.figure(figsize=(8,5))
sns.set(style="darkgrid",palette="summer")
df4["Author"].value_counts().plot.bar()
```

#### Dr. Seuss has most books(8) with 4.9 ratings.
