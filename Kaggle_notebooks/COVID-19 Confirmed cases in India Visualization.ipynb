### DATA COLLECTION

```python
import pandas as pd
df=pd.read_csv("covid_19_india.csv")
```

### DATA EXPLORATION & FEATURE ENGG

```python
df.isnull().mean()
```

```python
df.head()
```

```python
#dropping unimportant columns..
df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational","Cured","Deaths"],axis=1,inplace=True)
```

```python
df.head()
```

```python
#converting string column to date column
df['Date']= pd.to_datetime(df['Date'],dayfirst=True)#format="%m%d%Y")
```

### DATA VIZS

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### WESTBENGAL

```python
df2=df[df["State/UnionTerritory"]=="West Bengal"]
```

```python
plt.figure(figsize=(6,6))
sns.set(style="darkgrid",palette="winter_r")
plt.plot_date(df2["Date"],df2["Confirmed"],color='#056608')
plt.xticks(rotation=90)
plt.xlabel("Dates",fontdict={'fontsize':15})
plt.ylabel("Confirmed cases",fontdict={'fontsize':15})
plt.title("Growth Curve of Confirmed Corona Cases in West Bengal",fontdict={'fontsize':15})
```

```python
plt.figure(figsize=(8,6))
sns.barplot(df2["Date"],df2["Confirmed"])
plt.title("Confirmed Corona Cases in West Bengal",fontdict={'fontsize':25})
plt.xlabel("Passing of Time",fontdict={'fontsize':15})
plt.ylabel("Confirmed cases",fontdict={'fontsize':15})
plt.xticks([], [])
```

#### REPORT:

In our visualizations till the month of July there were no such addition of new cases. But from August when lockdown was relaxed gradually , there was a surge in new cases. This data is very much reliable because it is very much allingned with the real life.
